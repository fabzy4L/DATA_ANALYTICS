# CREATE Z SCORE TRANS FUNCTION

```python
def ztest(a):
    """ZSCORE TRANSFORMATION"""
    mean = np.mean(a, axis=0)  # Specify axis=0 to calculate column-wise mean
    std = np.std(a, axis=0)    # Specify axis=0 to calculate column-wise std deviation
    zscore = (a - mean) / std
    return zscore
```

# DEFINE DATA FOR TRANSFORMATION

```python
dftest1 = df_hnum
dftest2 = df_hdemo
```

# PERFORM TRANSFORMATION

```python
# Initialize an empty DataFrame to store the transformed data
transformed_df = pd.DataFrame()

# Iterate through the columns and perform the z-transformation
for column_name in columns_to_test:
    if column_name in dftest1.columns:
        column_data = dftest1[column_name]
        
        # Perform the z-transformation using your custom ztransformation function
        ztransformed_data = ztest(column_data)

        # Add the z-transformed data as a new column in the transformed_df
        transformed_df[column_name] = ztransformed_data

# Print or use the transformed DataFrame as needed
transformed_df

df_z = pd.concat([dftest2, transformed_df], axis=1)

df_z
```

# SHAPIRO TEST

```python
alpha = 0.05

# Initialize an empty list to store the results
shapiro_df = []

# Iterate through the columns and perform the Shapiro-Wilk test
for column_name_sw in columns_to_test:
    if column_name_sw in dftest1.columns:
        column_data_sw = dftest1[column_name_sw]
        statistic_sw, p_value_sw = stats.shapiro(column_data_sw)        
    
    # Determine normality based on the p-value
        if p_value_sw <= alpha:
            normality_sw = "**NOT** normally distributed"
        else:
            normality_sw = "Normally Distributed"

    # Store the result for the current column
        shapiro_df.append({
            "Column Name": column_name_sw,
            "Test Statistic": statistic_sw,
            "P-value": p_value_sw,
            "Normality": normality_sw
        })

# Create a DataFrame from the results
shapiro_hdf = pd.DataFrame(shapiro_df)

# Print or save the results as needed
shapiro_hdf
```
